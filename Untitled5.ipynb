```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')

#load scores generated previously for five departments
all_scores=pd.read_csv('all_scores_1.csv')

all_scores=all_scores.iloc[:,-5:].astype(str)

def rem_par(x):
    return float(''.join(c for c in x if c not in '()'))

#gather data into separate dataframes
perc_scores=pd.DataFrame({'svd_scores':[],'reg_scores':[],'sim_scores':[],'nmf_scores':[]})
for i in ['svd_scores','reg_scores','sim_scores','nmf_scores']:
    perc_scores[i]=all_scores[i].apply(lambda x: rem_par(x.split(',')[0]))
    
perc_rmsd=pd.DataFrame({'svd_scores':[],'reg_scores':[],'sim_scores':[],'nmf_scores':[]})
for i in ['svd_scores','reg_scores','sim_scores','nmf_scores']:
    perc_rmsd[i]=all_scores[i].apply(lambda x: rem_par(x.split(',')[1]))

overall_rmsd=pd.DataFrame({'svd_scores':[],'reg_scores':[],'sim_scores':[],'nmf_scores':[]})
for i in ['svd_scores','reg_scores','sim_scores','nmf_scores']:
    overall_rmsd[i]=all_scores[i].apply(lambda x: rem_par(x.split(',')[2]))
    
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  
                    textcoords="offset points",
                    ha='center', va='bottom')


#plot overall RMSD
fig, ax = plt.subplots(figsize=(10,4))

barWidth = 0.25

labels=list(all_scores['dept'])
labels=[i.replace('_',' ') for i in labels]

# set height of bar
bars1 = np.round(overall_rmsd['svd_scores'],2)
bars2 = np.round(overall_rmsd['reg_scores'],2)
bars3 = np.round(overall_rmsd['nmf_scores'],2)
 
# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
 
# Make the plot
rects1=ax.bar(r1, bars1,  width=barWidth, edgecolor='white', label='SVD')
rects2=ax.bar(r2, bars2,  width=barWidth, edgecolor='white', label='TFIDF + XGBoost')
rects3=ax.bar(r3, bars3,  width=barWidth, edgecolor='white', label='NMF')
 
# Add xticks on the middle of the group bars
ax.set_xticks(r2)
ax.set_xticklabels(labels,rotation=10)
ax.set_title('Overall RMSD')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

ax.set_ylim(0,2)
ax.legend(loc='upper left',ncol=3)
fig.tight_layout()
plt.savefig('overall_rmsd.png')

#plot RMSD of each user's predicted top 5%
fig, ax = plt.subplots(figsize=(10,4))

barWidth = 0.25
 
# set height of bar
bars1 = np.round(perc_rmsd['svd_scores'],2)
bars2 = np.round(perc_rmsd['reg_scores'],2)
bars3 = np.round(perc_rmsd['nmf_scores'],2)
 
# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
 
# Make the plot
rects1=ax.bar(r1, bars1,  width=barWidth, edgecolor='white', label='SVD')
rects2=ax.bar(r2, bars2,  width=barWidth, edgecolor='white', label='TFIDF + XGBoost')
rects3=ax.bar(r3, bars3,  width=barWidth, edgecolor='white', label='NMF')
 
# Add xticks on the middle of the group bars
ax.set_xticks(r2)
ax.set_xticklabels(labels,rotation=10)
ax.set_title('RMSD of each user\'s predicted top 5%')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

ax.set_ylim(0,2)
ax.legend(loc='upper left',ncol=3)
fig.tight_layout()
plt.savefig('top5_rmsd.png')

#plot average actual rating of each user's predicted top 5%
fig, ax = plt.subplots(figsize=(12,4))

barWidth = 0.2
 
# set height of bar
bars1 = np.round(perc_scores['svd_scores'],2)
bars2 = np.round(perc_scores['reg_scores'],2)
bars3 = np.round(perc_scores['nmf_scores'],2)
bars4 = np.round(perc_scores['sim_scores'],2)
 
# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]
 
# Make the plot
rects1=ax.bar(r1, bars1,  width=barWidth, edgecolor='white', label='SVD')
rects2=ax.bar(r2, bars2,  width=barWidth, edgecolor='white', label='TFIDF + XGBoost')
rects3=ax.bar(r3, bars3,  width=barWidth, edgecolor='white', label='NMF')
rects4=ax.bar(r4, bars4,  width=barWidth, edgecolor='white', label='TFIDF + Cosine Sim')

# Add xticks on the middle of the group bars
ax.set_xticks([x + barWidth/2 for x in r2])
ax.set_xticklabels(labels,rotation=10)
ax.set_title('Average actual ratings of each user\'s top predicted 5%')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
autolabel(rects4)

ax.set_ylim(3,5)
ax.legend(loc='upper left',ncol=4)
fig.tight_layout()
plt.savefig('top5_actual.png')
```
